### Inheritance:
- defining a new class with little or no modification to an existing class. The new class is called derived (or child) class and the one from which it inherits is called the base (or parent) class.

```python
from IPython.display import Image
```

### Multilevel inheritance

```python
Image(filename='/opsmnt/Projects/python_oop/Inh1.JPG')
```

```python
class Parent_Class:
    def action_method(self):
        print('From Parent class')

class Child_Class_Level_1_A(Parent_Class):
    pass

class Child_Class_Level_1_B():
    def action_method(self):
        print('From Child_Class_Level_1_B')

class Child_Class_Level_2(Child_Class_Level_1_A,Child_Class_Level_1_B):
    pass

cc_l2=Child_Class_Level_2()
cc_l2.action_method()
```

<b> Method resolution order

```python
Child_Class_Level_2.mro()
```

### Multilevel Dimond shape inheritance

```python
Image(filename='/opsmnt/Projects/python_oop/Inh2.JPG')
```

```python
class Parent_Class:
    def action_method(self):
        print('From Parent class')

class Child_Class_Level_1_A(Parent_Class):
    pass

class Child_Class_Level_1_B(Parent_Class):
    def action_method(self):
        print('From Child_Class_Level_1_B')

class Child_Class_Level_2(Child_Class_Level_1_A,Child_Class_Level_1_B):
    pass

cc_l2=Child_Class_Level_2()
cc_l2.action_method()
```

```python
Child_Class_Level_2.mro()
```
